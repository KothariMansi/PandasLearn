```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first": ["Mansi", "Manan", "Komal", "Sujal", 'Chris', np.nan, 'NA'],
    "last": ["Kothari", "Jain", "Kothari", "Jain", None, np.nan, "Missing"],
    "email": ["mansikothari46@gmail.com", "kotharimanan46@gmail.com", None, "sujalexample@gmail.com", 'chris@gamil.com', np.nan, 'NA'],
    "age": ['19', '16', np.nan, '18', '50', None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df.replace('NA', np.nan, inplace = True)
df.replace('Missing', np.nan, inplace = True)
```
